## Cargado del LLM (llama 3.1 -8b)

Se usa `llama-3.3-70b-versatile` porque soporta **tool calling** correctamente.

```python
llm  = ChatGroq(model="llama-3.3-70b-versatile", temperature=0)
```

##Creacion de herramientas para el agente

El agente cuenta con dos herramientas:

**buscar_reglamento(query):**
- Recupera los 8 chunks más relevantes de ChromaDB usando similitud coseno
- k=8 para capturar suficiente contexto en preguntas que abarcan varios artículos
- Se activa cuando el estudiante pregunta sobre normas o artículos del reglamento

**generar_resumen(texto):**
- Resume un texto en párrafos naturales y fluidos en español
- Sin listas ni bullet points, como si se lo explicaras a un compañero
- Se activa cuando el estudiante pide una explicación clara de un tema

```python
@tool
def buscar_reglamento(query: str) -> str:
    """
    Busca información relevante en el reglamento universitario.
    Usar cuando el estudiante haga preguntas sobre normas,
    artículos o políticas de la universidad.
    """
    docs = vectorstore.similarity_search(query, k=10)
    if not docs:
        return "No se encontró información relevante en el reglamento."
    return "\n\n".join([
        f"[Fragmento {i+1}]\n{doc.page_content}"
        for i, doc in enumerate(docs)
    ])
```

```python
@tool
def generar_resumen(texto: str) -> str:
    """
    Genera un resumen claro en español de un texto del reglamento.
    Usar cuando el estudiante pida explicación simple de un tema.
    """
    respuesta = llm.invoke(
        f"""Explica el siguiente fragmento del reglamento universitario
en un párrafo fluido y natural, como si se lo explicaras a un compañero estudiante.
Sin bullet points, sin listas, solo texto corrido y claro en español:

{texto}"""
    )
    return respuesta.content
```

```python
herramientas  = [buscar_reglamento, generar_resumen]
llm_con_tools = llm.bind_tools(herramientas)
tool_map      = {t.name: t for t in herramientas}
```

## Creacion de agente

Se evalúan preguntas de distinta complejidad para demostrar que el agente:
- Responde correctamente con información real del reglamento
- Cita artículos existentes sin inventar datos
- Responde en español de forma natural y fluida

```python
def consultar_agente(pregunta, verbose=True):
    print(f"\n{'='*60}")
    print(f"ESTUDIANTE: {pregunta}")
    print(f"{'='*60}\n")

    t0 = time.time()
    contexto = buscar_reglamento.invoke({"query": pregunta})
    if verbose:
        print(f"  → Herramienta: buscar_reglamento")
        print(f"    Resultado: {contexto[:200]}...\n")
    mensajes = [HumanMessage(content=f"""Eres un asistente universitario amigable.
Responde en español, en párrafos naturales y fluidos.
Usa ÚNICAMENTE la información del contexto proporcionado.
Si la respuesta no está en el contexto, di exactamente: "No encontré esa información en el reglamento."
NO inventes artículos ni información que no esté en el contexto.

Contexto del reglamento:
{contexto}

Pregunta del estudiante: {pregunta}""")]

    respuesta = llm.invoke(mensajes)
    tiempo = time.time() - t0

    print(f"RESPUESTA FINAL ({tiempo:.1f}s):")
    print(respuesta.content)
    print(f"\n{'='*60}\n")
    return respuesta.content
```

## Pruebas de agente:

#### Pregunta: ¿Qué pasa si desapruebo el mismo curso tres veces?

```python
consultar_agente("¿Qué pasa si desapruebo el mismo curso tres veces?")
```

#### Pregunta: ¿Que articulos hablan sobre sanciones y que dicen, selecciona 3

```python
consultar_agente("¿Que articulos hablan sobre sanciones y que dicen, selecciona 3?")
```
